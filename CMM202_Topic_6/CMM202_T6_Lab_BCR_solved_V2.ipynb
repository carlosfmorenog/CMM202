```python
# Importing pandas
import pandas as pd
# This code makes charts appear in the notebook in the first run
%matplotlib inline 
```

First, we import the data

```python
df = pd.read_csv('https://www.dropbox.com/s/wkvlf6hclrlrr46/city_populations_updated.csv?raw=1')
df
```

We only include years above a certain limit (otherwise it would take ages to plot!)

```python
df2 = df[df['Year']>1991]
df2.reset_index()
```

Now we pivot the table to comply with the way that `bcr` likes data!

```python
# Pivot data
df3 = df2.pivot(index='Year',
                          columns='City',
                          values=['Population'])
df3 = df3['Population']# If we don't do this, then the word population will appear next to the city in the BCR
df3
```

Finally, we use the function contained in the `bcr` package to do the plot. As of 05/02/2022, there is no option to group the countries by continent (also, LA doesn't go down!).

```python
import warnings
warnings.filterwarnings('ignore')
import bar_chart_race as bcr
bcr.bar_chart_race(df=df3, filename=None,
                   n_bars=12,
                   perpendicular_bar_func='median', 
                   title='Most Populous Cities in the World')
```
